```python
import pandas as pd

def sort_csv(input_file, output_file, sort_column, ascending=True):
    # Read the CSV file with semicolon delimiter
    df = pd.read_csv(input_file, delimiter=';')

    # Remove extra spaces in column names
    df.columns = df.columns.str.strip()

    # Check if the sort column exists
    if sort_column not in df.columns:
        print(f"Error: Column '{sort_column}' not found in CSV. Available columns: {df.columns.tolist()}")
        return

    # Sort the DataFrame by the specified column
    df_sorted = df.sort_values(by=sort_column, ascending=ascending)

    # Save the sorted DataFrame to a new CSV file
    df_sorted.to_csv(output_file, index=False)
    print(f"Sorted CSV saved as {output_file}")

# Display column names to verify the correct column for sorting
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/HAI Security Dataset.csv', delimiter=';')
df.columns = df.columns.str.strip()
print("Available columns:", df.columns.tolist())

# Example usage with the correct column name
sort_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/HAI Security Dataset.csv',
         '/content/drive/MyDrive/Colab Notebooks/dataset hai/ordered.csv',
         sort_column='time')  # Replace 'time' with the desired column

# Load and display the sorted CSV to verify
sorted_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/ordered.csv')
print(sorted_df.head())

```

# **3.1	Integrated Anomaly Detection Ensemble (IADE)**
# **3.1.1	Support Vector Machine (SVM)**
# **3.1.2	Random Forest**
# **3.1.3	Autoencoders**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report

# Load the smart meter data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Autoencoder model for anomaly detection
input_dim = X_train.shape[1]
encoding_dim = 14

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation='sigmoid')(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer=Adam(learning_rate=0.009), loss='mse')

# Train autoencoder on normal data
normal_data = X_train[y_train == 0]
autoencoder.fit(normal_data, normal_data, epochs=100, batch_size=12, shuffle=True, validation_split=0.2)

# Reconstruction error
reconstructions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - reconstructions, 2), axis=1)
threshold = np.percentile(mse, 95)

# Detect anomalies
anomaly_labels = (mse > threshold).astype(int)

# SVM for classification
svm = SVC(kernel='rbf', C=1, gamma='auto')
svm.fit(X_train, y_train)
svm_predictions = svm.predict(X_test)

# Random Forest for ensemble decision
rf = RandomForestClassifier(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)
rf_predictions = rf.predict(X_test)

# Ensemble Decision: Majority Voting
def ensemble_decision(autoencoder_labels, svm_preds, rf_preds):
    combined = np.vstack((autoencoder_labels, svm_preds, rf_preds)).T
    return np.array([np.argmax(np.bincount(row)) for row in combined])

iade_predictions = ensemble_decision(anomaly_labels, svm_predictions, rf_predictions)

# Evaluation
classification_rep = classification_report(y_test, iade_predictions, output_dict=True)

# Confusion matrix
cm = confusion_matrix(y_test, iade_predictions)

# Extract values from the confusion matrix
TN, FP, FN, TP = cm.ravel()

# Calculate the metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity) if (precision + sensitivity) != 0 else 0
mcc = ((TP * TN) - (FP * FN)) / np.sqrt((FP + TP) * (FN + TP) * (TN + FP) * (TN + FN)) if all([FP + TP, FN + TP, TN + FP, TN + FN]) != 0 else 0
npv = TN / (FN + TN) if (FN + TN) != 0 else 0
fpr = FP / (TN + FP) if (TN + FP) != 0 else 0
fnr = FN / (TP + FN) if (TP + FN) != 0 else 0

# Save predictions and metrics to a DataFrame
metrics = {
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Sensitivity': [sensitivity],
    'Specificity': [specificity],
    'F1-Score': [f1_score],
    'MCC': [mcc],
    'NPV': [npv],
    'FPR': [fpr],
    'FNR': [fnr],
}

metrics_df = pd.DataFrame(metrics)

# Save the predictions and metrics to a CSV file
predictions_df = pd.DataFrame({
    'True Labels': y_test,
    'Predictions': iade_predictions
})

# Save the metrics and predictions to CSV
metrics_df.to_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/IADE_metrics.csv', index=False)
predictions_df.to_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/IADE_predictions.csv', index=False)
print("Metrics and predictions saved to CSV.")
predictions_df



```

# **AdaptiveSlidingModeController**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import time

# Load the smart meter data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Adaptive Sliding Mode Controller (ASMC) Algorithm
class AdaptiveSlidingModeController:
    def __init__(self, system_dynamics, control_limits, lambda_1=1.0, lambda_2=1.0):
        self.system_dynamics = system_dynamics
        self.u_min, self.u_max = control_limits
        self.lambda_1 = lambda_1
        self.lambda_2 = lambda_2
        self.sigma = 0.0
        self.alpha = 1.0

    def sliding_surface(self, x):
        s = x[0] + self.lambda_1 * x[1]
        return s

    def control_law(self, x, s):
        u = -self.lambda_1 * np.sign(s) - self.lambda_2 * self.alpha * s
        u = np.clip(u, self.u_min, self.u_max)
        return u

    def adaptation_law(self, s, u, alpha):
        alpha_dot = -self.lambda_2 * np.sign(s) * (u - self.alpha * s)
        alpha += alpha_dot * 0.01
        return alpha

    def simulate(self, initial_state, time_span, dt=0.01):
        x = np.array(initial_state)
        sim_time = np.arange(0, time_span, dt)  # Avoid conflict with the time module
        states = []
        control_inputs = []

        for t in sim_time:
            s = self.sliding_surface(x)
            u = self.control_law(x, s)
            control_inputs.append(u)
            states.append(x)
            dx = self.system_dynamics(x, u)
            x = x + dx * dt
            self.alpha = self.adaptation_law(s, u, self.alpha)

        return np.array(states), np.array(control_inputs), sim_time

# Example system: 2nd-order system (mass-spring-damper model)
def system_dynamics(x, u):
    m = 1.0  # mass
    b = 0.5  # damping coefficient
    k = 1.0  # spring constant
    x1, x2 = x
    dx1 = x2
    dx2 = (u - b * x2 - k * x1) / m
    return np.array([dx1, dx2])

# Initialize the ASMC controller
control_limits = [-10, 10]
asmc_controller = AdaptiveSlidingModeController(system_dynamics, control_limits)

# Simulate the system
initial_state = [0.0, 0.0]
time_span = 10.0
states, control_inputs, sim_time = asmc_controller.simulate(initial_state, time_span)

# Plot results
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(sim_time, states[:, 0], label='Position')
plt.plot(sim_time, states[:, 1], label='Velocity')
plt.xlabel('Time [s]')
plt.ylabel('State')
plt.legend(loc='best')
plt.title('State Evolution over Time')

plt.subplot(2, 1, 2)
plt.plot(sim_time, control_inputs, label='Control Input (Force)')
plt.xlabel('Time [s]')
plt.ylabel('Control Input')
plt.legend(loc='best')
plt.title('Control Input over Time')

plt.tight_layout()
plt.show()

# Measure the computation time
start_time = time.time()
# End the timer after the computations
end_time = time.time()
computation_time = end_time - start_time

# Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity) if (precision + sensitivity) != 0 else 0
mcc = ((TP * TN) - (FP * FN)) / np.sqrt((FP + TP) * (FN + TP) * (TN + FP) * (TN + FN)) if all([FP + TP, FN + TP, TN + FP, TN + FN]) else 0
npv = TN / (FN + TN) if (FN + TN) != 0 else 0
fpr = FP / (TN + FP) if (TN + FP) != 0 else 0
fnr = FN / (TP + FN) if (TP + FN) != 0 else 0

# Computational Efficiency
total_correct_predictions = TP + TN
computational_efficiency = total_correct_predictions / computation_time if computation_time > 0 else 0

# Save metrics to a DataFrame
metrics = {
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Sensitivity': [sensitivity],
    'Specificity': [specificity],
    'F1-Score': [f1_score],
    'MCC': [mcc],
    'NPV': [npv],
    'FPR': [fpr],
    'FNR': [fnr],
    'Computational Efficiency': [computational_efficiency],
    'Computation Time (seconds)': [computation_time],
}

metrics_df = pd.DataFrame(metrics)

# Save metrics to a CSV file
metrics_csv_path = '/content/drive/MyDrive/Colab Notebooks/dataset hai/metrics1.csv'
metrics_df.to_csv(metrics_csv_path, index=False)

print(f"Performance metrics saved to {metrics_csv_path}")
print(metrics_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, matthews_corrcoef
import time

# Load the smart meter data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=13)

# Model-Free Adaptive Control (MFAC) Algorithm
class ModelFreeAdaptiveControl:
    def __init__(self, control_limits, lambda_1=1.0, lambda_2=1.0):
        self.u_min, self.u_max = control_limits
        self.lambda_1 = lambda_1
        self.lambda_2 = lambda_2
        self.u = 0.0  # Initial control input

    def control_law(self, error):
        # Adapt control input based on error dynamics
        u = -self.lambda_1 * np.sign(error) - self.lambda_2 * error
        u = np.clip(u, self.u_min, self.u_max)
        self.u = u
        return u

# Initialize MFAC controller
control_limits = [-10, 10]
mfac_controller = ModelFreeAdaptiveControl(control_limits)

# Start measuring the computation time
start_time = time.time()

# Train a classifier (Random Forest)
classifier = RandomForestClassifier(n_estimators=10, random_state=0)

# Fit the classifier to the training data
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

# Calculate error (difference between actual and predicted labels)
error = y_test - y_pred

# Use MFAC to adapt control based on error dynamics (simulating an adaptive process)
control_inputs = np.array([mfac_controller.control_law(e) for e in error])

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
sensitivity = recall_score(y_test, y_pred)
specificity = (confusion_matrix(y_test, y_pred)[0, 0] / (confusion_matrix(y_test, y_pred)[0, 0] + confusion_matrix(y_test, y_pred)[0, 1]))
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
npv = tn / (fn + tn)
fpr = fp / (tn + fp)
fnr = fn / (tp + fn)

# End measuring the computation time
end_time = time.time()

# Calculate computation time and computational efficiency
computation_time = end_time - start_time
computational_efficiency = X_train.shape[0] / computation_time

# Print performance metrics
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Sensitivity (Recall): {sensitivity:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"MCC: {mcc:.2f}")
print(f"NPV: {npv:.2f}")
print(f"FPR: {fpr:.2f}")
print(f"FNR: {fnr:.2f}")

# Print control inputs (MFAC)
print(f"Control Inputs (MFAC): {control_inputs}")

# Print computation metrics
print(f"Computation Time: {computation_time:.4f} seconds")
print(f"Computational Efficiency: {computational_efficiency:.4f} computations per second")
```

```python
import time
import random
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, matthews_corrcoef
from sklearn.ensemble import RandomForestClassifier

# Load the smart meter data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=13)

# Define the OvercurrentRelay class
class OvercurrentRelay:
    def __init__(self, threshold, reset_time=10):
        self.threshold = threshold  # Overcurrent threshold in Amps
        self.reset_time = reset_time  # Time to wait before resetting the relay (in seconds)
        self.status = "Normal"  # Initial relay status

    def monitor_current(self):
        # Simulate the current measurement (in Amps)
        current = random.uniform(0, 2 * self.threshold)  # Random current between 0 and twice the threshold
        return current

    def check_overcurrent(self, current):
        """Check if the current exceeds the overcurrent threshold."""
        if current > self.threshold:
            self.trigger_relay()
        else:
            self.reset_relay()

    def trigger_relay(self):
        """Simulate the relay being triggered due to overcurrent."""
        self.status = "Tripped"
        print(f"Overcurrent detected! Relay status: {self.status}")
        # Simulate the time the relay stays tripped before being reset
        time.sleep(self.reset_time)
        self.reset_relay()

    def reset_relay(self):
        """Reset the relay back to normal state."""
        if self.status == "Tripped":
            print(f"Relay reset after {self.reset_time} seconds. Relay status: Normal")
            self.status = "Normal"

    def run(self):
        """Run the monitoring and checking process."""
        while True:
            # Monitor the current in the system
            current = self.monitor_current()
            print(f"Current: {current:.2f} A")

            # Check if current exceeds the threshold
            self.check_overcurrent(current)

            # Wait for a while before next measurement (simulate time delay)
            time.sleep(1)

# Set the threshold current for overcurrent protection (e.g., 10 Amps)
threshold_current = 10

# Create the overcurrent relay object
relay = OvercurrentRelay(threshold=threshold_current)

# Train the model
model = RandomForestClassifier()

# Start time for training
start_time = time.time()

# Fit the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
sensitivity = recall_score(y_test, y_pred)
specificity = (confusion_matrix(y_test, y_pred)[0, 0] / (confusion_matrix(y_test, y_pred)[0, 0] + confusion_matrix(y_test, y_pred)[0, 1]))
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
npv = tn / (fn + tn)
fpr = fp / (tn + fp)
fnr = fn / (tp + fn)

# End time for training
end_time = time.time()

# Compute computation time
computation_time = end_time - start_time

# Estimate number of operations (this is an approximation; adjust based on your knowledge of the model)
# Random Forest operation approximation: for each decision tree, the operations per sample can be assumed to be constant.
# Here, we assume a constant number of operations for simplicity.
num_operations = 1e7  # Adjust this based on the model complexity and number of operations per sample

# Compute computational efficiency
computational_efficiency = num_operations / computation_time if computation_time > 0 else 0

# Print performance metrics
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Sensitivity (Recall): {sensitivity:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"MCC: {mcc:.2f}")
print(f"NPV: {npv:.2f}")
print(f"FPR: {fpr:.2f}")
print(f"FNR: {fnr:.2f}")
print(f"Computation Time: {computation_time:.4f} seconds")
print(f"Computational Efficiency: {computational_efficiency:.4f} computations per second")

```

# **RCoDEMS**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import time

class RCoDEMS:
    def __init__(self, data, num_clusters=3, max_iter=100, tol=1e-4):
        """
        R-CoDEMS: Robust Constrained Dynamic Model-based Segmentation for time-series data
        :param data: Input signal (time-series or one-dimensional data for segmentation)
        :param num_clusters: Number of clusters
        :param max_iter: Maximum iterations for convergence
        :param tol: Convergence tolerance
        """
        self.data = data
        self.num_clusters = num_clusters
        self.max_iter = max_iter
        self.tol = tol

        self.clusters = np.zeros(data.shape, dtype=int)  # Store cluster assignments
        self.centroids = np.random.rand(num_clusters)  # Random initial centroids

    def fit(self):
        """
        Fit the model to segment the data into clusters
        """
        prev_centroids = np.copy(self.centroids)
        for iteration in range(self.max_iter):
            # Step 1: Assign data points to nearest centroid
            self.assign_clusters()

            # Step 2: Update centroids
            self.update_centroids()

            # Check for convergence
            centroid_change = np.linalg.norm(self.centroids - prev_centroids)
            if centroid_change < self.tol:
                print(f"Converged after {iteration + 1} iterations")
                break
            prev_centroids = np.copy(self.centroids)

    def assign_clusters(self):
        """
        Assign each data point to the nearest centroid
        """
        distances = np.abs(self.data[:, np.newaxis] - self.centroids)  # Compute distance to centroids
        cluster_assignments = np.argmin(distances, axis=1)
        self.clusters = cluster_assignments

    def update_centroids(self):
        """
        Update centroids based on current cluster assignments
        """
        for i in range(self.num_clusters):
            cluster_points = self.data[self.clusters == i]
            if len(cluster_points) > 0:
                self.centroids[i] = np.mean(cluster_points)

    def plot_results(self):
        """
        Plot the segmented time-series data
        """
        plt.figure(figsize=(10, 6))
        plt.plot(self.data, label='Original Data')
        plt.scatter(range(len(self.data)), self.data, c=self.clusters, cmap='viridis', label='Clustered Data')
        plt.title("Segmentation of Time-Series Data")
        plt.xlabel("Time")
        plt.ylabel("Signal Value")
        plt.legend()
        plt.show()

# Load the smart meter data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=13)

# Example: Apply R-CoDEMS to the smart meter data

# Assuming 'attack' is your relevant column for analysis
time_series_data = data['attack'].values  # Replace 'attack' with the column name for time series

# Create R-CoDEMS instance and fit model
rcode_ms = RCoDEMS(time_series_data, num_clusters=3, max_iter=100, tol=1e-4)

# Measure computation time
start_time = time.time()

# Fit the model
rcode_ms.fit()

# End the timer after the computations
end_time = time.time()
computation_time = end_time - start_time

# Plot the results
rcode_ms.plot_results()

# Define confusion matrix metrics (example)
TP = 50  # True positives (need actual data for calculation)
TN = 30  # True negatives (need actual data for calculation)
FP = 10  # False positives (need actual data for calculation)
FN = 5   # False negatives (need actual data for calculation)

# Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity) if (precision + sensitivity) != 0 else 0
mcc = ((TP * TN) - (FP * FN)) / np.sqrt((FP + TP) * (FN + TP) * (TN + FP) * (TN + FN)) if all([FP + TP, FN + TP, TN + FP, TN + FN]) else 0
npv = TN / (FN + TN) if (FN + TN) != 0 else 0
fpr = FP / (TN + FP) if (TN + FP) != 0 else 0
fnr = FN / (TP + FN) if (TP + FN) != 0 else 0

# Computational Efficiency
total_correct_predictions = TP + TN
computational_efficiency = total_correct_predictions / computation_time if computation_time > 0 else 0

# Save metrics to a DataFrame
metrics = {
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Sensitivity': [sensitivity],
    'Specificity': [specificity],
    'F1-Score': [f1_score],
    'MCC': [mcc],
    'NPV': [npv],
    'FPR': [fpr],
    'FNR': [fnr],
    'Computational Efficiency': [computational_efficiency],
    'Computation Time (seconds)': [computation_time],
}

metrics_df = pd.DataFrame(metrics)
print(metrics_df)

```

# **egc_lstm Algorithm**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report, matthews_corrcoef
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Conv1D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import time

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=13)

# Ensure the number of features is divisible by the chosen timesteps
num_timesteps = 31  # This should divide the number of features, here 62 // 2 = 31
features_per_timestep = 2  # Adjust this to ensure division

X_train_reshaped = X_train.reshape((X_train.shape[0], num_timesteps, features_per_timestep))  # 31 timesteps, 2 features each
X_test_reshaped = X_test.reshape((X_test.shape[0], num_timesteps, features_per_timestep))  # 31 timesteps, 2 features each

# Build the EGC-LSTM model
def build_egc_lstm(input_shape):
    model = Sequential()

    # Conv1D Layer (to process sequential data)
    model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(Dropout(0.2))  # Dropout to prevent overfitting

    # LSTM Layer (for sequence processing)
    model.add(LSTM(50, dropout=0.2))

    # Fully Connected Layer (Dense)
    model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

    # Compile the model
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Build and train the model
input_shape = X_train_reshaped.shape[1:]  # Use reshaped input shape
egc_lstm_model = build_egc_lstm(input_shape)

# Train the model and measure computation time
start_time = time.time()
history = egc_lstm_model.fit(X_train_reshaped, y_train, epochs=100, batch_size=32, validation_data=(X_test_reshaped, y_test))
end_time = time.time()

# Evaluate the model
loss, accuracy = egc_lstm_model.evaluate(X_test_reshaped, y_test)

# Predict the test data
y_pred = (egc_lstm_model.predict(X_test_reshaped) > 0.5).astype(int)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
TN, FP, FN, TP = conf_matrix.ravel()

# Additional metrics
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
fpr = FP / (FP + TN) if (FP + TN) > 0 else 0
fnr = FN / (FN + TP) if (FN + TP) > 0 else 0
mcc = matthews_corrcoef(y_test, y_pred)
npv = TN / (TN + FN) if (TN + FN) > 0 else 0

# Computational metrics
computation_time = end_time - start_time
computational_efficiency = 1 / computation_time if computation_time > 0 else 0

# Print the results
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall (Sensitivity): {recall}")
print(f"F1-Score: {f1_score}")
print(f"Specificity: {specificity}")
print(f"False Positive Rate (FPR): {fpr}")
print(f"False Negative Rate (FNR): {fnr}")
print(f"Matthews Correlation Coefficient (MCC): {mcc}")
print(f"Negative Predictive Value (NPV): {npv}")
print(f"Computation Time: {computation_time} seconds")
print(f"Computational Efficiency: {computational_efficiency} computations per second")


```

# **Deep Graph Learning**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, matthews_corrcoef, precision_score, recall_score, f1_score
from torch_geometric.data import Data
from sklearn.model_selection import train_test_split
import time

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset hai/cleaned_file.csv')

# Identify and exclude non-numeric columns (e.g., date columns)
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data_numeric = data.drop(columns=non_numeric_columns)

# Separate features and target
X = data_numeric.drop('attack', axis=1)
y = data['attack']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Convert to tensor
X_tensor = torch.tensor(X_scaled, dtype=torch.float)
y_tensor = torch.tensor(y.values, dtype=torch.long)

# Since the data is not graph-structured, let's assume a fully connected graph with nodes as data points
num_nodes = X_scaled.shape[0]
edge_index = torch.combinations(torch.arange(num_nodes), r=2).t()

# Create Data object for PyTorch Geometric
data = Data(x=X_tensor, edge_index=edge_index, y=y_tensor)

# Split the data into training and test sets
train_indices, test_indices = train_test_split(np.arange(len(y_tensor)), test_size=0.2, random_state=42)

# Create train_mask and test_mask
train_mask = torch.zeros(len(y_tensor), dtype=torch.bool)
test_mask = torch.zeros(len(y_tensor), dtype=torch.bool)

train_mask[train_indices] = True
test_mask[test_indices] = True

# Assign the masks to the data object
data.train_mask = train_mask
data.test_mask = test_mask

# Define the dgl model (Deep Graph Learning)
class GCN(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(GCN, self).__init__()
        # Graph Convolution Layer 1
        self.conv1 = GCNConv(in_channels, 16)
        # Graph Convolution Layer 2
        self.conv2 = GCNConv(16, out_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        # Apply first graph convolution
        x = F.relu(self.conv1(x, edge_index))
        # Apply second graph convolution
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)  # Classification output

# Instantiate the model
model = GCN(in_channels=X_tensor.shape[1], out_channels=len(data.y.unique()))

# Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

# Metrics calculation functions
def calculate_metrics(y_true, y_pred):
    conf_matrix = confusion_matrix(y_true, y_pred)
    TN, FP, FN, TP = conf_matrix.ravel()

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
    fpr = FP / (FP + TN) if (FP + TN) > 0 else 0
    fnr = FN / (FN + TP) if (FN + TP) > 0 else 0
    mcc = matthews_corrcoef(y_true, y_pred)
    npv = TN / (TN + FN) if (TN + FN) > 0 else 0

    return precision, recall, f1, specificity, fpr, fnr, mcc, npv

# Function to calculate computation time and computational efficiency
def calculate_computational_efficiency(start_time, num_operations):
    end_time = time.time()
    computation_time = end_time - start_time
    computational_efficiency = num_operations / computation_time if computation_time > 0 else 0
    return computation_time, computational_efficiency

# Training loop with added metrics
def train():
    model.train()
    optimizer.zero_grad()

    # Start timing the computation
    start_time = time.time()

    out = model(data)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()

    # Estimate number of operations per forward pass (adjust as needed)
    num_operations = 1e7  # Replace with an approximation of operations per pass
    computation_time, computational_efficiency = calculate_computational_efficiency(start_time, num_operations)

    return loss.item(), computation_time, computational_efficiency

# Testing loop with added metrics
def test():
    model.eval()
    with torch.no_grad():
        out = model(data)
        pred = out.argmax(dim=1)

        # Get the prediction and ground truth labels for the test set
        y_true = data.y[data.test_mask].cpu().numpy()
        y_pred = pred[data.test_mask].cpu().numpy()

        # Calculate accuracy
        correct = (pred[data.test_mask] == data.y[data.test_mask]).sum()
        accuracy = int(correct) / int(data.test_mask.sum())

        # Calculate other metrics
        precision, recall, f1, specificity, fpr, fnr, mcc, npv = calculate_metrics(y_true, y_pred)

    return accuracy, precision, recall, f1, specificity, fpr, fnr, mcc, npv

# Training the model
for epoch in range(200):
    loss, computation_time, computational_efficiency = train()
    if epoch % 10 == 0:
        accuracy, precision, recall, f1, specificity, fpr, fnr, mcc, npv = test()
        print(f'Epoch: {epoch}, Loss: {loss:.4f}, Accuracy: {accuracy:.4f}')
        print(f'Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}, Specificity: {specificity:.4f}')
        print(f'FPR: {fpr:.4f}, FNR: {fnr:.4f}, MCC: {mcc:.4f}, NPV: {npv:.4f}')
        print(f"Computation Time: {computation_time:.4f} seconds")
        print(f"Computational Efficiency: {computational_efficiency:.4f} computations per second")

```
